```python
import matplotlib.pyplot as plt
import numpy as np

# Data
costs = [32.3725, 28.9227, 34.531, 28.9048, 33.7645, 36.0855]
revenues = [40.7904, 28.616, 41.0928, 28.7056, 36.0976, 47.88]
profits = [8.4179, -0.3067, 6.5618, -0.1992, 2.3331, 11.7945]
labels = ['Case 1', 'Case 2', 'Case 3', 'Case 4', 'Case 5', 'Case 6']

# Create figure
plt.figure(figsize=(10, 6))

# Define the width of each bar
bar_width = 0.2

# Create an array for the positions of the bars
index = np.arange(len(labels))

# Plot bars with shifted positions to avoid overlap
plt.bar(index, costs, width=bar_width, label='Average Cost', color='blue')
plt.bar(index + bar_width, revenues, width=bar_width, label='Average Revenue', color='green')
plt.bar(index + 2 * bar_width, profits, width=bar_width, label='Average Profit', color='red')

# Add title and axis labels
plt.title('Comparison of Average Costs, Revenues, and Profits')
plt.xlabel('Case')
plt.ylabel('Amount')

# Add tick positions and labels
plt.xticks(index + bar_width, labels)

# Add legend
plt.legend()

# Show the figure
plt.tight_layout()
plt.show()
```
